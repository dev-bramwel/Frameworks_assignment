```python
#import libraries
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

# import data from dataset
df = pd.read_csv('metadata_sample.csv')
print("Data loaded successfully")
print("Begin data cleaning")
```

```python
# data cleaning
#get a snapshot of the data
print ("The following is some information about the data‚úÖ\n")
print(df.info()) #generate an informatory description about the data
print("\nHere are the first 10 rowsüí≤\n")
print(df.head()) # show the first 5 rows
print ("\nThis is a descriptive summaryüìä\n")
print(df.describe())
```

```python
#Analysing the data
print("Begin data analysis\n1Ô∏è‚É£ total records")
#Get total records
print(f"Total Records: {len(df)}")
#Group by time in years
print("\n2Ô∏è‚É£Grouping in years")
df['year'] = pd.to_datetime(df['publish_time']).dt.year
grouped = df.groupby('year')
print(grouped.size())
# List all unique values in 'source_x' and their counts
print("\n3Ô∏è‚É£Listing all sources")
source_counts = df['source_x'].value_counts()
print(source_counts)
# Select rows from a specific source
print("\n4Ô∏è‚É£Grouping data with same source")
selected_source = 'Elsevier'
rows_from_source = df[df['source_x'] == selected_source]
print(rows_from_source)
```

```python
#Visualization
print("The following is the visualization of the data on the frequency of each source\n")
# Get counts of each source
source_counts = df['source_x'].value_counts()

# Plot bar chart
plt.figure(figsize=(10,6))
source_counts.plot(kind='bar')
plt.title('Frequency of Each Source')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nA line graph showing how many studies were published yearly\n")
# line graph showing the studies pubished in each year
df['year'] = pd.to_datetime(df['publish_time']).dt.year

# Count rows per year
year_counts = df['year'].value_counts().sort_index()

# Plot line graph
plt.figure(figsize=(10,6))
plt.plot(year_counts.index, year_counts.values, marker='o')
plt.title('Number of Rows Published Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
